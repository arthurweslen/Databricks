#### Importando Pyspark

```
import pyspark
from pyspark.sql import SparkSession
  
```

#### Configuração Pyspark

```
spark = SparkSession.builder.appName('escolher_nome').getOrCreate()
print('SparkSessioon.bilder criada e com sucesso! \n'+ str(spark))
```

#### Criando um dicionário

```
# list  of college data with  dictionary

data = [{'id': 1, 'fruta': 'maçã'},
        {'id': 2, 'fruta': 'banana'},
        {'id': 3, 'fruta': 'pera'},
        {'id': 4, 'fruta': 'abacaxi'},
        {'id': 5, 'fruta': 'laranja'},
        {'id': 6, 'fruta': 'limão'} ]
  
```

#### Criando um Dataframe

```
# creating a dataframe
dataframe = spark.createDataFrame(data)
  
# show data frame
dataframe.show()
```
